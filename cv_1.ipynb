```python
import matplotlib.pyplot as plt
import seaborn as sns
import pydataset as pyd

cars = pyd.data('mtcars')
```

# 1. Histogram

```python


plt.figure(figsize=(15, 3))
sns.histplot(data=cars['hp'], bins = 20, color='red', linewidth=2, edgecolor="white")
plt.ylabel("Count")
plt.xlabel("Horsepower")
plt.title("Histogram of horsepower")
plt.show()

plt.figure(figsize=(15, 3))
sns.histplot(data=cars['mpg'], bins= 20, color='black', edgecolor="white")
plt.ylabel("Count")
plt.xlabel("Miles per Galon")
plt.title("Histogram of Miles per Galon")

plt.figure(figsize=(15, 3))
sns.histplot(data=cars['wt'], bins = 20, color='cyan', edgecolor='white')
plt.ylabel("Count")
plt.xlabel("Weight(t)")
plt.title("Histogram of weight")
plt.show()

```

# 2. Scatterplot

```python
plt.figure(figsize=(15,3))
sns.scatterplot(data=cars, x = 'hp', y = 'mpg', size=10, legend=False)
plt.xlabel("Horsepower")
plt.ylabel("Miles per Gallon")
plt.title("Horsepower vs Miles per Gallon")
plt.show()

plt.figure(figsize=(15,3))
sns.scatterplot(data=cars, x = 'mpg', y = 'wt', size=10, legend=False)
plt.xlabel("Miles per Gallon")
plt.ylabel("Weight")
plt.title("Miles per Gallon vs Weight")
plt.show()

plt.figure(figsize=(15,3))
sns.scatterplot(data=cars, x = 'wt', y = 'hp', size=10, legend=False)
plt.xlabel("Weight")
plt.ylabel("Horsepower")
plt.title("Weight vs Horsepower")
plt.show()
```

# 3. Boxplots

```python
plt.figure(figsize=(15,3))
sns.boxplot(cars, x= 'cyl', y='mpg', color='green')
plt.ylabel("Miles per Gallon")
plt.xlabel("Cylinders")
plt.title("Miles per Galon grouped by Cylinders")
plt.show()

plt.figure(figsize=(15,3))
sns.boxplot(cars, x= 'cyl', y='hp', color='purple')
plt.ylabel("Horsepower")
plt.xlabel("Cylinders")
plt.title("Horsepower grouped by Cylinders")
plt.show()

plt.figure(figsize=(15,3))
sns.boxplot(cars, x= 'cyl', y='wt', color='pink')
plt.ylabel("Weight")
plt.xlabel("Cylinders")
plt.title("Weight grouped by Cylinders")
plt.show()

```

# 4. Reflection

## Histograms
- Most cars have around 100-200 Hp
- Most cars run around 15-20 Mpg
- Most cars weight around 3-3.5 T

## Scatterplots
- Less Hp means more Mpg
- Less weight means more Mpg
- Cars around 3.5 T have more Hp  

## Boxplots
- Less cylinders means more Mpg
- More cylinders mostly means more Hp
- More cylinders mostly means more W
